```python
import numpy as np
import matplotlib.pyplot as plt
```

The code below defines the $20$-th order Wilkinson polynomial. 

```python
roots = np.array([i for i in range(1, 21)], dtype = int)
p = np.polynomial.Polynomial.fromroots(roots)
```

This is how you evaluate it at a point. The added `print` ensures that the result is printed, otherwise only the last result in a cell will be printed.

```python
print(p(1.5))
```
